```python
df_vehicle_type_count = df.groupby('Electric Vehicle Type').size()
print(df_vehicle_type_count)
```

```python
import matplotlib.pyplot as plt
def datalabel(x,y):
  for i in range(len(x)):
    plt.text(i,y[i],y[i])

l1 = ['grey','orange']
l2 = df_vehicle_type_count.index
plt.figure(figsize=(10,5))
bars = plt.bar(df_vehicle_type_count.index,df_vehicle_type_count.values, color=l1,label=l2)
plt.title('Electric Vehicle Type ')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Total vehicle')
plt.legend(title='Type of vehicle')
datalabel(df_vehicle_type_count.index,df_vehicle_type_count.values)
plt.show()
```
